Dataset Description

The dataset has records of 299 patients who suffered a clinical heart failure and the information collected on 13 attributes during their follow-up period. The 13 attributes mentioned in the dataset is as follows


Age: Age of the patient (in years)\
Anaemia: If or if not there was a decrease in haemoglobin or red blood cells (Boolean values)\
Sex: Male or Female (Binary)\
High Blood Pressure: If or if not the patient has Hypertension (Boolean)\
Diabetes: Whether the patient suffers from diabetes or not (Boolean)\
Smoking: Whether the patient smokes or not (Boolean)\
Creatinine Phosphokinase (CPK): Level of the enzyme CPK in the blood (in mcg/L)\
Ejection Fraction: The percentage of blood leaving the heart at each contraction (Percentage)\
Platelets: Amount of platelets present in the blood (in kiloplatelets/mL)\
Serum Creatinine: The level of Serum Creatinine in the blood (in mg/dL)\
Serum Sodium: The level of Serum Sodium in the blood (in mEq/L)\
Time: The follow-up period (in days)\
Death Event: Whether or not the patient died in the follow-up period (Boolean)\
\
mcg/L: Micrograms per litre, mEq/L: Milliequivalents per litre

After this Linear regression is performed to find the attributes most prevalent to the death event.
Prediction is also done for some example values

```python
import pandas
from sklearn import linear_model

x = df[['age', 'anaemia', 'creatinine_phosphokinase' , 'diabetes' , 
        'ejection_fraction' , 'high_blood_pressure' , 'platelets' , 'serum_creatinine' , 'serum_sodium' , 
        'sex' , 'smoking' , 'time']]
y = df['DEATH_EVENT']
```

As it is the Target Parameter Death Event is created as a separate list and the parameters as another one

```python
regr = linear_model.LinearRegression()
regr.fit(x, y)

print(regr.coef_)
```

Regression is performed and the coefficients are shown as follows. These coefficients aren't very helpful for forming a correlation as the data isn't normalized but can be found suitable for prediction. A prediction example is shown below

```python
predictdeath = regr.predict([[62,0,61,1,38,1,155000,1.1,143,1,1,270]])

print(predictdeath)
```

The produced predicted value is rounded off to be in line with the Boolean values for Death Event

```python
xyz = np.round(predictdeath,0)
print(abs(xyz))
```
